```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
image = cv2.imread('Desktop/Udacity/hough.jpg')
#plt.imshow(image)

%matplotlib inline 
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
print(image_copy.shape)
```

```python
#convert image to grey scale 

grey= cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

#apply canny edge detection 

low_threshold = 0
high_threshold =100
edges = cv2.Canny(grey, low_threshold, high_threshold)
plt.imshow(edges, cmap='gray')
```

```python
#find lines using the Hough Transform 

rho =1 #resolution 
theta = np.pi/180
threshold = 60 #minimum number of hough space intersection
min_length_line = 10
max_line_gap = 2

lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_length_line, max_line_gap )

#make a copy of the image 

line_image = np.copy(image_copy)

#iterate over 'lines' and draw the lines on the line_image 
for line in lines: 
    for x1,y1,x2,y2 in line: 
        cv2.line(line_image, (x1,y1),(x2,y2), (255, 0,0), 5) # 255 represents the colour and 5 repreents the line width 
        
plt.imshow(line_image)
```
